```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
timestep = 20
file = np.array(df)
file_X = file[:, 1:]
file_y = file[:, 0:1]

print(file_X.shape)
print(file_y.shape)
def prepare_inputs(X_train, X_test):
    ohe = OneHotEncoder(sparse = False)
    #ohe.fit(X_train[:,5:])
    X_train_enc = ohe.fit_transform(X_train[:,:])
    #X_train_enc = np.asarray(X_train_enc)
    #X = []
    #for i in range(len(X_train_enc)):
    #    X.append(X_train_enc[i,:])
    #X= np.array(X)
    #print(X.shape)
#     print(X_train_enc.dtype)
#     X_train_enc = np.array(X_train_enc)
#     print(X_train_enc.shape)
#     X_train_enc = np.pad(X_train_enc, (5,0),'constant', constant_values=0)
    #X_test_enc = ohe.transform(X_test)
    X_train_enc = np.array(X_train_enc)
#     X_train_enc = np.c_[X_train[:,:5], X_train_enc]
    print(X_train_enc[0])
    print(X_train_enc.shape)
    return X_train_enc

file_X = prepare_inputs(file_X, file_y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(file_X, file_y, test_size=0.33, random_state=42)
```

```python
DENSE_CONCAT = 512
top_words = 500
max_review_length = 50
embedding_vecor_length = 32
weight_decay = 0.0000
dropout=0.2
Embedding_outdim = 128
output_dim=1
ADD_ON_LAYERS = True
from keras.layers import Dense,LSTM,Activation,Dropout,BatchNormalization,Input,Embedding,Reshape,Concatenate, GRU
from keras.models import Sequential,Model
```

```python
X_train2= np.asarray(X_train2).astype(np.float32)
X_test2= np.asarray(X_test2).astype(np.float32)

y_train= np.asarray(y_train).astype(np.float32)

X_train= np.asarray(X_train).astype(np.float32)
X_test= np.asarray(X_test).astype(np.float32)

input1 = Input(shape=(X_train2.shape[1],), dtype='float32')
input12 = Embedding(top_words, embedding_vecor_length, input_length=max_review_length)(input1)
input13 = Dropout(0.2)(input12)
input14 = LSTM(100)(input13)
input15 = Dropout(0.2)(input14)
input16 = Dense(64, activation='relu')(input15)
text_vec = Dense(128, activation='sigmoid')(input16)

input2 = Input(shape=(X_train.shape[1],), dtype='float32')
input21 = Dense(128, activation='relu', kernel_initializer='he_normal')(input2)
input22 = Dense(64, activation='relu')(input21)
Poi = Dense(128, activation='sigmoid')(input22)

level_3 = Concatenate()([text_vec,Poi])
print(text_vec.shape)
print(Poi.shape)
print(level_3.shape)
```

```python
model_in = Dense(DENSE_CONCAT,
    kernel_regularizer=regularizers.l2(weight_decay),
    activation='relu')(level_3)

model_in = Dense(units=int(DENSE_CONCAT/2),
    kernel_regularizer=regularizers.l2(weight_decay),
    activation=None)(model_in)
if ADD_ON_LAYERS:
    model_in = BatchNormalization()(model_in)

model_in = Activation('relu')(model_in)
model_in = Dropout(dropout)(model_in)

model_in = Dense(units=int(DENSE_CONCAT/8),
    kernel_regularizer=regularizers.l2(weight_decay),
    activation=None)(model_in)
if ADD_ON_LAYERS:
    model_in = BatchNormalization()(model_in)

model_in = Activation('relu')(model_in)

main_output = Dense(output_dim, activation='sigmoid')(model_in)

model = Model(inputs=[input1,input2], outputs=main_output)
print(model.summary())
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```
